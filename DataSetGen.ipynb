# ***Getting Image path to CSV files***

### ***Saving image path to CSV files***

```python
ImgPath = 'asl'
DataSavingPath = 'H:\Research\SignLanguage\MyData'
foldernames = os.listdir(ImgPath)

for i, value in enumerate(foldernames):
    filenames = os.listdir(f'{ImgPath}/{value}')
    # print(value)
    with open(f'{DataSavingPath}\{value}.csv', 'w') as f:
        # using csv.writer method from CSV package
        write = csv.writer(f)
        for name in range(len(filenames)):
            write.writerow([filenames[name]] + [f'{value}'])
```

### ***Loading CSV files to dataframe***

```python
#read the path
file_path = "DataSet"
#list all the files from the directory
file_list = os.listdir(file_path)
file_list
```

```python
def preProcess(img):
    
    # img = cv2.resize(img, (200, 200))
    img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # img = tf.keras.utils.normalize(img, axis = 1) #normalizing data

    return img
```

```python
from matplotlib.colors import NoNorm

img_1 = (mpimg.imread("asl/0/0_10_rotate_1.jpeg"))
img_2 = (mpimg.imread("asl/4/4_1_rotate_1.jpeg"))
img_3 = (mpimg.imread("asl/h/h_1_rotate_1.jpeg"))

img1 = preProcess(img_1)
img2 = preProcess(img_2)
img3 = preProcess(img_3)

retval, threshold = cv2.threshold(img_1, 12, 255, cv2.THRESH_BINARY)

# img2 = preProcess(img)
# img2 = np.expand_dims(img2, axis=0)
# print(img2.shape)
fig=plt.figure(figsize=(15,15))
fig.add_subplot(1,3,1)
plt.title("Preprocessed Sign of digit 0")
plt.imshow(img1, cmap='gray', norm=NoNorm())

fig.add_subplot(1,3,2)
plt.title("Preprocessed sign for digit 4")
plt.imshow(img2, cmap='gray', norm=NoNorm())

fig.add_subplot(1,3,3)
plt.title("Preprocessed Sign for alphabet h")
plt.imshow(img3, cmap='gray', norm=NoNorm())
plt.show()
```
